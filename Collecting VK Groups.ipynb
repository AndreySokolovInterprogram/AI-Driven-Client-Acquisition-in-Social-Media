```python
#–ü–æ–ª—É—á–∞–µ–º —Å–ª–æ–≤–∞—Ä—å. –ö–ª—é—á - id –≥—Ä—É–ø–ø—ã, –∑–Ω–∞—á–µ–Ω–∏–µ - —Å–ø–∏—Å–æ–∫ –ø–æ–ª—å–∑–æ–≤–∞–µ—Ç–µ–ª–π —ç—Ç–æ–π –≥—Ä—É–ø–ø—ã
from time import sleep
groups = [17700656, 99946133, 142786786, 89373035, 171840666]
usersOfAllGroups = dict()
for grp in groups:
  group_users = []
  offset = 0

  response = vk.groups.getMembers(group_id = grp, offset=0)
  len_1 = response["count"]
  # len_1 = 3000

  while offset < len_1:
    response = vk.groups.getMembers(group_id = grp, offset=offset)
    group_users += response["items"]
    offset += 1000
    sleep(1)
  usersOfAllGroups[grp] = group_users
```
